```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

### Navigate to Result Page

```python
ecom_website = 'https://www.geekbuying.com/'
```

```python
# initialize Chrome Driver
driver = webdriver.Chrome('C:/webdrivers/chromedriver.exe')

# open website
driver.get(ecom_website)

# maximize window
driver.maximize_window()
```

```python
click_x = driver.find_element_by_xpath('(//span[@class="close_btn line_close close_sub"])[1]')
click_x.click()
```

```python
search = driver.find_element_by_xpath('//div[@class="search_input"]/input[2]')
search.click()

```

```python
search.send_keys('e bike')
```

```python
search.send_keys(Keys.ENTER)
```

### Start The Web Scraping Process

#### Locators

```python
# 2. dann Dataframe kreieren
df_bike = pd.DataFrame(columns=['Product Name', 'Price New', 'Price Old', 'Likes'])

# 1. Locators setzen
product_name = driver.find_elements_by_xpath('//li[@class="searchResultItem"]/div/div[2]/a')
product_price_new = driver.find_elements_by_xpath('//li[@class="searchResultItem"]/div/div[3]')
product_price_old = driver.find_elements_by_xpath('//li[@class="searchResultItem"]/div/div[4]')
likes = driver.find_elements_by_xpath('//span[@class="favite_count"]')

# 3. in Dataframe appenden
for i in range(len(product_name)):

    df_bike =df_bike.append({'Product Name':product_name[i].text, 'Price New': product_price_new[i].text,
                            'Price Old': product_price_old[i].text, 'Likes': likes[i].text}, ignore_index=True)
```
